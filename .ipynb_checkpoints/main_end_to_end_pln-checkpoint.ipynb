## 0.3 Criação da base intermediária

Nesta etapa será criada uma base intermediária com uma coluna textual unificada chamada `text_raw`, que reúne título e texto da avaliação, quando ambos estiverem disponíveis, servindo para padronizar o ponto de partida das próximas seções do pipeline.

```python
df_base = df.copy()

for col in ["review_title", "review_text"]:
    if col in df_base.columns:
        df_base[col] = df_base[col].fillna("").astype(str)

if "review_title" in df_base.columns and "review_text" in df_base.columns:
    df_base["text_raw"] = (
        df_base["review_title"].str.strip()
        + ". "
        + df_base["review_text"].str.strip()
    )
elif "review_text" in df_base.columns:
    df_base["text_raw"] = df_base["review_text"].str.strip()
else:
    raise ValueError("Nenhuma coluna textual esperada foi encontrada.")

df_base["text_raw"] = (
    df_base["text_raw"]
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)

df_base["doc_id"] = range(len(df_base))

display(df_base[["doc_id", "text_raw"]].head())
```

## 1.1 Carregamento da base intermediária

A base utilizada nesta seção é `data/interim/repro_base.csv`, que contém a coluna `text_raw`, que reúne o título e o corpo de cada avaliação. A partir dela serão criadas novas colunas processadas para uso nas próximas etapas do projeto.

```python
from pathlib import Path
import os
import re
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from collections import Counter

BASE_DIR = Path.cwd()

DATA_RAW_DIR = BASE_DIR / "data" / "raw"
DATA_INTERIM_DIR = BASE_DIR / "data" / "interim"
DATA_PROCESSED_DIR = BASE_DIR / "data" / "processed"

OUTPUT_FIGURES_DIR = BASE_DIR / "outputs" / "figures"
OUTPUT_TABLES_DIR = BASE_DIR / "outputs" / "tables"

for directory in [DATA_PROCESSED_DIR, OUTPUT_FIGURES_DIR, OUTPUT_TABLES_DIR]:
    directory.mkdir(parents=True, exist_ok=True)

base_path = DATA_INTERIM_DIR / "repro_base.csv"

df = pd.read_csv(base_path)

print("Base carregada com sucesso.")
print("Dimensões:", df.shape)

display(df.head())
```

## 1.2 Tokenização de sentenças e palavras

A tokenização é a etapa que divide o texto em unidades menores. Neste projeto são usadas duas formas:

- tokenização de sentenças, para estimar a estrutura dos documentos;
- tokenização de palavras, para preparar o vocabulário usado nas próximas etapas.

A tokenização será feita com NLTK, utilizando recursos compatíveis com textos em português.

```python
import nltk

nltk.download("punkt", quiet=True)
nltk.download("punkt_tab", quiet=True)
nltk.download("stopwords", quiet=True)
nltk.download("rslp", quiet=True)

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import RSLPStemmer
```

```python
def tokenize_sentences_pt(text):
    text = "" if pd.isna(text) else str(text)
    return sent_tokenize(text, language="portuguese")

def tokenize_words_pt(text):
    text = "" if pd.isna(text) else str(text)
    return word_tokenize(text, language="portuguese")

df["sentences_raw"] = df["text_raw"].apply(tokenize_sentences_pt)
df["tokens_raw"] = df["text_raw"].apply(tokenize_words_pt)

df["n_sentences"] = df["sentences_raw"].apply(len)
df["n_tokens_raw"] = df["tokens_raw"].apply(len)

display(df[["doc_id", "text_raw", "sentences_raw", "tokens_raw", "n_sentences", "n_tokens_raw"]].head(3))
```

## 1.3 Normalização textual

A normalização textual reduz variações superficiais do texto. Nesta etapa são aplicadas as seguintes transformações:

- conversão para letras minúsculas;
- remoção de URLs;
- remoção de e-mails;
- remoção de caracteres que não são letras;
- redução de espaços duplicados.

Os acentos foram preservados porque o corpus está em português e a preservação de acentuação ajuda a manter a qualidade linguística da lemmatização com spaCy.

```python
def normalize_text(text):
    text = "" if pd.isna(text) else str(text)
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", " ", text)
    text = re.sub(r"\S+@\S+", " ", text)
    text = re.sub(r"[^a-záàâãéèêíïóôõöúçñ\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["text_norm"] = df["text_raw"].apply(normalize_text)
df["tokens_norm"] = df["text_norm"].apply(tokenize_words_pt)
df["n_tokens_norm"] = df["tokens_norm"].apply(len)

display(df[["text_raw", "text_norm", "tokens_norm"]].head(5))
```

## 1.4 Remoção de stopwords

Stopwords são palavras muito frequentes que geralmente carregam pouco conteúdo semântico isoladamente, como artigos, preposições e pronomes. Além da lista padrão de stopwords em português do NLTK, foram adicionadas stopwords customizadas comuns em avaliações informais, como abreviações e marcas de oralidade.

Termos centrais do domínio, como `produto`, `entrega`, `preço`, `marca` e `loja`, foram preservados neste momento, pois podem ser importantes para a análise de tópicos, classificação e construção do grafo de conhecimento.

```python
stopwords_pt = set(stopwords.words("portuguese"))

custom_stopwords = {
    "pra", "pro", "pros", "pras",
    "vc", "vcs", "voce", "voçê",
    "ta", "tá", "ne", "né",
    "ai", "aí", "q", "pq",
    "tb", "tambem", "também"
}

stopwords_final = stopwords_pt.union(custom_stopwords)

def remove_stopwords_tokens(tokens):
    clean_tokens = []
    for token in tokens:
        token = token.lower().strip()
        if token.isalpha() and token not in stopwords_final and len(token) > 2:
            clean_tokens.append(token)
    return clean_tokens

df["tokens_no_stop"] = df["tokens_norm"].apply(remove_stopwords_tokens)
df["text_no_stop"] = df["tokens_no_stop"].apply(lambda tokens: " ".join(tokens))

display(df[["text_norm", "tokens_no_stop", "text_no_stop"]].head(5))
```

## 1.5 Comparação entre stemming e lemmatização

Stemming e lemmatização são técnicas para reduzir variações de palavras. O stemming reduz palavras a radicais aproximados, em português, isso pode reduzir bastante o vocabulário, mas frequentemente gera formas pouco interpretáveis. Já a lemmatização reduz palavras ao seu lema, preservando melhor a forma linguística. Por isso, tende a ser mais adequada quando a interpretação dos termos é importante, como neste projeto.

```python
stemmer = RSLPStemmer()

def apply_stemming(tokens):
    return [stemmer.stem(token) for token in tokens]

df["tokens_stem"] = df["tokens_no_stop"].apply(apply_stemming)
df["text_stem"] = df["tokens_stem"].apply(lambda tokens: " ".join(tokens))

display(df[["tokens_no_stop", "tokens_stem", "text_stem"]].head(5))
```

## 1.6 Lemmatização e POS tagging com spaCy

Nesta etapa, o spaCy é usado para duas tarefas:

- lemmatização dos tokens;
- identificação da classe gramatical dos termos, conhecida como POS tagging.

O POS tagging ajuda a caracterizar linguisticamente o corpus, permitindo observar se as avaliações são compostas principalmente por substantivos, adjetivos, verbos ou outras classes. Em avaliações de produtos, substantivos e adjetivos tendem a ser especialmente relevantes, pois indicam objetos avaliados e qualificações atribuídas pelos consumidores.

```python
import spacy

nlp = spacy.load("pt_core_news_sm")

print("Modelo spaCy carregado.")
print("Pipeline:", nlp.pipe_names)
```

```python
lemma_tokens_by_doc = []
pos_records = []

texts = df["text_norm"].fillna("").tolist()
doc_ids = df["doc_id"].tolist()

for doc_id, doc in zip(doc_ids, nlp.pipe(texts, batch_size=128, n_process=1)):
    lemmas = []
    
    for token in doc:
        lemma = token.lemma_.lower().strip()
        
        lemma_parts = lemma.split()
        
        if (
            token.is_alpha
            and not token.is_stop
            and len(lemma) > 2
            and lemma not in stopwords_final
            and not any(part in stopwords_final for part in lemma_parts)
            and token.pos_ not in {"ADP", "DET", "PRON", "CCONJ", "SCONJ"}
        ):
            lemmas.append(lemma)
            
            pos_records.append({
                "doc_id": doc_id,
                "token": token.text.lower(),
                "lemma": lemma,
                "pos": token.pos_
            })
    
    lemma_tokens_by_doc.append(lemmas)

df["tokens_lemma"] = lemma_tokens_by_doc
df["text_lemma"] = df["tokens_lemma"].apply(lambda tokens: " ".join(tokens))

pos_df = pd.DataFrame(pos_records)

display(df[["text_norm", "tokens_lemma", "text_lemma"]].head(5))
display(pos_df.head())
```

```python
pos_counts = (
    pos_df["pos"]
    .value_counts()
    .reset_index()
)

pos_counts.columns = ["pos", "frequencia"]

display(pos_counts)

pos_counts.to_csv(
    OUTPUT_TABLES_DIR / "section_1_pos_distribution.csv",
    index=False,
    encoding="utf-8"
)
```

## 1.7 Impacto do pré-processamento no vocabulário

Para avaliar o impacto das decisões de pré-processamento, serão comparados os tamanhos de vocabulário em quatro versões:

1. tokens normalizados;
2. tokens sem stopwords;
3. tokens com stemming;
4. tokens com lemmatização.

Essa comparação permite verificar quanto cada etapa reduz o vocabulário e qual estratégia preserva melhor a interpretabilidade dos termos.

```python
def calculate_vocab_stats(token_series, label):
    total_tokens = sum(len(tokens) for tokens in token_series)
    vocab = set(token for tokens in token_series for token in tokens)
    
    return {
        "representacao": label,
        "total_tokens": total_tokens,
        "tamanho_vocabulario": len(vocab),
        "media_tokens_por_documento": round(total_tokens / len(token_series), 2)
    }

vocab_stats = pd.DataFrame([
    calculate_vocab_stats(df["tokens_norm"], "tokens_normalizados"),
    calculate_vocab_stats(df["tokens_no_stop"], "sem_stopwords"),
    calculate_vocab_stats(df["tokens_stem"], "stemming"),
    calculate_vocab_stats(df["tokens_lemma"], "lemmatizacao")
])

display(vocab_stats)

vocab_stats.to_csv(
    OUTPUT_TABLES_DIR / "section_1_vocab_impact.csv",
    index=False,
    encoding="utf-8"
)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(vocab_stats["representacao"], vocab_stats["tamanho_vocabulario"])
plt.title("Impacto do pré-processamento no tamanho do vocabulário")
plt.xlabel("Representação")
plt.ylabel("Quantidade de termos únicos")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_1_vocab_impact.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 1.9 Decisão técnica sobre a representação textual

Após a comparação entre stemming e lemmatização, a versão lematizada foi adotada como principal representação textual para as próximas etapas do pipeline, porque a lemmatização preserva melhor a interpretabilidade dos termos em português. Embora o stemming reduza o vocabulário de forma mais agressiva, ele pode gerar radicais artificiais e menos compreensíveis, dificultando a interpretação de tópicos, termos frequentes, resultados de busca textual e relações no grafo de conhecimento. Assim, o stemming foi utilizado como comparação metodológica, enquanto a lemmatização será usada como base principal para representação vetorial, análise de tópicos, classificação e visualizações posteriores.

# Seção 2 - Representação vetorial e busca textual

Esta seção transforma os textos pré-processados em representações numéricas para análise computacional.

Serão utilizadas duas abordagens principais:

1. Bag-of-Words com `CountVectorizer`;
2. TF-IDF com `TfidfVectorizer`, utilizando unigramas e bigramas.

A partir dessas representações, será construído um mecanismo simples de busca textual por similaridade. O usuário poderá informar uma consulta textual e o sistema retornará os documentos mais similares do corpus, acompanhados de seus respectivos scores.

Nesta seção, a busca será testada com três consultas diferentes relacionadas ao domínio de avaliações de e-commerce.

## 2.1 Carregamento da base pré-processada

A base utilizada nesta seção é `data/processed/repro_preprocessed.csv`, gerada na Seção 1, e a coluna principal utilizada será `text_lemma`, que contém os textos normalizados, sem stopwords e lematizados.

```python
from pathlib import Path
import os
import re
import json
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from collections import Counter

BASE_DIR = Path.cwd()

DATA_PROCESSED_DIR = BASE_DIR / "data" / "processed"
OUTPUT_FIGURES_DIR = BASE_DIR / "outputs" / "figures"
OUTPUT_TABLES_DIR = BASE_DIR / "outputs" / "tables"
OUTPUT_MODELS_DIR = BASE_DIR / "outputs" / "models"

for directory in [OUTPUT_FIGURES_DIR, OUTPUT_TABLES_DIR, OUTPUT_MODELS_DIR]:
    directory.mkdir(parents=True, exist_ok=True)

processed_path = DATA_PROCESSED_DIR / "repro_preprocessed.csv"

df = pd.read_csv(processed_path)

print("Base pré-processada carregada com sucesso.")
print("Dimensões:", df.shape)

display(df.head())
```

```python
required_columns = ["doc_id", "text_raw", "text_lemma"]

missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"Colunas obrigatórias ausentes: {missing_columns}")

df["text_lemma"] = df["text_lemma"].fillna("").astype(str)
df["text_raw"] = df["text_raw"].fillna("").astype(str)

df_vec = df[df["text_lemma"].str.len() > 0].copy()

print("Documentos disponíveis para vetorização:", len(df_vec))
display(df_vec[["doc_id", "text_raw", "text_lemma"]].head())
```

## 2.2 Bag-of-Words com CountVectorizer

A primeira representação vetorial utilizada é o Bag-of-Words, nessa abordagem, cada documento é representado pela frequência dos termos presentes no vocabulário, a ordem das palavras é desconsiderada, mas a presença e a repetição dos termos são preservadas, sendo uma abordagem simples e interpretável, servindo como uma representação de referência para comparação com TF-IDF.

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

bow_vectorizer = CountVectorizer(
    max_features=5000,
    min_df=3
)

bow_matrix = bow_vectorizer.fit_transform(df_vec["text_lemma"])

print("Matriz Bag-of-Words criada.")
print("Formato da matriz:", bow_matrix.shape)
print("Tamanho do vocabulário:", len(bow_vectorizer.vocabulary_))
```

## 2.3 TF-IDF com unigramas e bigramas

A segunda representação utilizada é o TF-IDF. Diferentemente do Bag-of-Words, o TF-IDF atribui maior peso a termos que são importantes em um documento, mas não aparecem de forma excessivamente comum em todo o corpus.

Nesta etapa, o TF-IDF foi configurado com unigramas e bigramas. Os unigramas capturam termos individuais, enquanto os bigramas permitem preservar expressões compostas relevantes, como "entrega atrasada", "custo benefício" ou "produto defeito".

```python
tfidf_vectorizer = TfidfVectorizer(
    max_features=5000,
    min_df=3,
    ngram_range=(1, 2)
)

tfidf_matrix = tfidf_vectorizer.fit_transform(df_vec["text_lemma"])

print("Matriz TF-IDF criada.")
print("Formato da matriz:", tfidf_matrix.shape)
print("Tamanho do vocabulário:", len(tfidf_vectorizer.vocabulary_))
```

## 2.4 Comparação entre as representações

As duas representações serão comparadas quanto ao tamanho do vocabulário, formato da matriz, quantidade de valores não nulos e densidade. A densidade indica a proporção de posições preenchidas na matriz, matrizes textuais geralmente são esparsas, pois cada documento utiliza apenas uma pequena parte do vocabulário total.

```python
def matrix_density(matrix):
    total_positions = matrix.shape[0] * matrix.shape[1]
    return matrix.nnz / total_positions

vectorization_summary = pd.DataFrame([
    {
        "representacao": "Bag-of-Words",
        "vetorizador": "CountVectorizer",
        "ngram_range": "1,1",
        "documentos": bow_matrix.shape[0],
        "termos": bow_matrix.shape[1],
        "valores_nao_nulos": bow_matrix.nnz,
        "densidade": matrix_density(bow_matrix)
    },
    {
        "representacao": "TF-IDF",
        "vetorizador": "TfidfVectorizer",
        "ngram_range": "1,2",
        "documentos": tfidf_matrix.shape[0],
        "termos": tfidf_matrix.shape[1],
        "valores_nao_nulos": tfidf_matrix.nnz,
        "densidade": matrix_density(tfidf_matrix)
    }
])

display(vectorization_summary)

vectorization_summary.to_csv(
    OUTPUT_TABLES_DIR / "section_2_vectorization_summary.csv",
    index=False,
    encoding="utf-8"
)
```

```python
plt.figure(figsize=(7, 4))
plt.bar(vectorization_summary["representacao"], vectorization_summary["termos"])
plt.title("Quantidade de termos nas representações vetoriais")
plt.xlabel("Representação")
plt.ylabel("Número de termos")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_2_vectorization_terms_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 2.6 Motor de busca textual por similaridade

O mecanismo de busca textual será construído com base em similaridade por cosseno.

O fluxo da busca é:

1. Receber uma consulta textual.
2. Aplicar normalização e lematização na consulta.
3. Transformar a consulta no mesmo espaço vetorial do corpus.
4. Calcular a similaridade entre a consulta e todos os documentos.
5. Retornar os documentos mais similares com seus respectivos scores.

O motor será testado tanto com Bag-of-Words quanto com TF-IDF, permitindo comparar o impacto das duas representações nos resultados da busca.

```python
import nltk
import spacy

from nltk.corpus import stopwords
from sklearn.metrics.pairwise import cosine_similarity

nltk.download("stopwords", quiet=True)

nlp = spacy.load("pt_core_news_sm")

stopwords_pt = set(stopwords.words("portuguese"))

custom_stopwords = {
    "pra", "pro", "pros", "pras",
    "vc", "vcs", "voce", "voçê",
    "ta", "tá", "ne", "né",
    "ai", "aí", "q", "pq",
    "tb", "tambem", "também"
}

stopwords_final = stopwords_pt.union(custom_stopwords)

def normalize_query(text):
    text = "" if pd.isna(text) else str(text)
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", " ", text)
    text = re.sub(r"\S+@\S+", " ", text)
    text = re.sub(r"[^a-záàâãéèêíïóôõöúçñ\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def preprocess_query(text):
    text_norm = normalize_query(text)
    doc = nlp(text_norm)
    
    lemmas = []
    for token in doc:
        lemma = token.lemma_.lower().strip()
        if token.is_alpha and lemma not in stopwords_final and len(lemma) > 2:
            lemmas.append(lemma)
    
    return " ".join(lemmas)

def search_similar_documents(query, vectorizer, matrix, data, top_n=5, representation_name="TF-IDF"):
    query_processed = preprocess_query(query)
    query_vector = vectorizer.transform([query_processed])
    
    similarities = cosine_similarity(query_vector, matrix).ravel()
    top_indices = similarities.argsort()[::-1][:top_n]
    
    results = data.iloc[top_indices].copy()
    results["rank"] = range(1, top_n + 1)
    results["score"] = similarities[top_indices]
    results["consulta_original"] = query
    results["consulta_processada"] = query_processed
    results["representacao"] = representation_name
    
    selected_cols = [
        "representacao",
        "consulta_original",
        "consulta_processada",
        "rank",
        "score",
        "doc_id",
        "text_raw"
    ]
    
    optional_cols = [
        "product_name",
        "product_brand",
        "site_category_lv1",
        "overall_rating",
        "polarity"
    ]
    
    selected_cols += [col for col in optional_cols if col in results.columns]
    
    return results[selected_cols]
```

## 2.7 Testes com três consultas

O motor de busca será testado com três consultas representativas do domínio de avaliações de e-commerce:

1. `produto chegou com defeito`;
2. `entrega atrasada`;
3. `ótimo custo benefício`.

Essas consultas foram escolhidas porque representam situações comuns em avaliações de produtos: defeito, problema logístico e avaliação positiva de valor percebido.

```python
queries = [
    "produto chegou com defeito",
    "entrega atrasada",
    "ótimo custo benefício"
]

all_search_results = []

for query in queries:
    bow_results = search_similar_documents(
        query=query,
        vectorizer=bow_vectorizer,
        matrix=bow_matrix,
        data=df_vec,
        top_n=5,
        representation_name="Bag-of-Words"
    )
    
    tfidf_results = search_similar_documents(
        query=query,
        vectorizer=tfidf_vectorizer,
        matrix=tfidf_matrix,
        data=df_vec,
        top_n=5,
        representation_name="TF-IDF"
    )
    
    all_search_results.append(bow_results)
    all_search_results.append(tfidf_results)

search_results_df = pd.concat(all_search_results, ignore_index=True)

display(search_results_df)

search_results_df.to_csv(
    OUTPUT_TABLES_DIR / "section_2_search_results.csv",
    index=False,
    encoding="utf-8"
)
```

```python
for query in queries:
    print("=" * 100)
    print("Consulta:", query)
    print("=" * 100)
    
    query_results = search_results_df[
        (search_results_df["consulta_original"] == query)
        & (search_results_df["representacao"] == "TF-IDF")
    ].copy()
    
    for _, row in query_results.iterrows():
        print(f"\nRank {row['rank']} | Score: {row['score']:.4f} | doc_id: {row['doc_id']}")
        
        if "product_name" in row.index:
            print("Produto:", row["product_name"])
        
        if "site_category_lv1" in row.index:
            print("Categoria:", row["site_category_lv1"])
        
        if "polarity" in row.index:
            print("Polaridade:", row["polarity"])
        
        print("Texto:", str(row["text_raw"])[:500], "...")
```

# Seção 3 - Modelagem, classificação e análise de sentimento

Esta seção extrai significado estruturado do corpus por meio de classificação supervisionada de sentimento.

Como o corpus RePro possui rótulos de polaridade, a estratégia adotada foi treinar modelos supervisionados para classificar avaliações como positivas ou negativas.

Serão utilizadas duas frentes principais:

1. Regressão Logística;
2. SVM Linear.

Essas abordagens foram escolhidas porque ambas são adequadas para classificação textual com representações TF-IDF, apresentam boa interpretabilidade operacional e são compatíveis com execução em computador pessoal. A avaliação será feita com métricas de classificação, incluindo precision, recall, F1-score e matriz de confusão.

## 3.1 Carregamento da base pré-processada

A base utilizada nesta seção é `data/processed/repro_preprocessed.csv`, a coluna `text_lemma` será usada como entrada textual, contendo a versão normalizada, sem stopwords e lematizada das avaliações e a coluna `polarity` será utilizada para construir o rótulo supervisionado de sentimento.

```python
from pathlib import Path
import os
import ast
import json
import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

BASE_DIR = Path.cwd()

DATA_PROCESSED_DIR = BASE_DIR / "data" / "processed"
OUTPUT_FIGURES_DIR = BASE_DIR / "outputs" / "figures"
OUTPUT_TABLES_DIR = BASE_DIR / "outputs" / "tables"
OUTPUT_MODELS_DIR = BASE_DIR / "outputs" / "models"

for directory in [OUTPUT_FIGURES_DIR, OUTPUT_TABLES_DIR, OUTPUT_MODELS_DIR]:
    directory.mkdir(parents=True, exist_ok=True)

processed_path = DATA_PROCESSED_DIR / "repro_preprocessed.csv"

df = pd.read_csv(processed_path)

print("Base pré-processada carregada com sucesso.")
print("Dimensões:", df.shape)

display(df.head())
```

```python
required_columns = ["doc_id", "text_raw", "text_lemma", "polarity"]

missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"Colunas obrigatórias ausentes: {missing_columns}")

df["text_lemma"] = df["text_lemma"].fillna("").astype(str)
df["text_raw"] = df["text_raw"].fillna("").astype(str)

display(df[required_columns].head())
```

## 3.2 Construção do rótulo de sentimento

A coluna `polarity` do corpus contém anotações de polaridade. Para esta etapa, o problema foi formulado como uma classificação binária entre avaliações positivas e negativas.

Avaliações com polaridade mista, neutra ou indefinida foram removidas desta etapa para manter uma tarefa supervisionada mais clara e coerente, já que o objetivo principal desta seção é comparar classificadores supervisionados em uma tarefa de sentimento com rótulos bem definidos.

```python
def parse_list_like(value):
    if isinstance(value, (list, tuple, set)):
        return list(value)
    
    if pd.isna(value):
        return []
    
    value_str = str(value).strip()
    
    try:
        parsed = ast.literal_eval(value_str)
        if isinstance(parsed, (list, tuple, set)):
            return list(parsed)
        return [parsed]
    except Exception:
        return [value_str]

def simplify_polarity(value):
    values = {str(item).upper().strip() for item in parse_list_like(value)}
    
    has_positive = "POSITIVO" in values
    has_negative = "NEGATIVO" in values
    has_neutral = "NEUTRO" in values
    
    if has_positive and not has_negative and not has_neutral:
        return "positivo"
    
    if has_negative and not has_positive and not has_neutral:
        return "negativo"
    
    if has_neutral and not has_positive and not has_negative:
        return "neutro"
    
    if has_positive and has_negative:
        return "misto"
    
    return "outro"

df["sentiment_label"] = df["polarity"].apply(simplify_polarity)

label_distribution = (
    df["sentiment_label"]
    .value_counts(dropna=False)
    .reset_index()
)

label_distribution.columns = ["classe", "quantidade"]

display(label_distribution)

label_distribution.to_csv(
    OUTPUT_TABLES_DIR / "section_3_sentiment_label_distribution_full.csv",
    index=False,
    encoding="utf-8"
)
```

```python
df_model = df[
    (df["sentiment_label"].isin(["positivo", "negativo"]))
    & (df["text_lemma"].str.len() > 0)
].copy()

print("Dimensões da base de modelagem:", df_model.shape)

class_distribution = (
    df_model["sentiment_label"]
    .value_counts()
    .reset_index()
)

class_distribution.columns = ["classe", "quantidade"]
class_distribution["proporcao"] = class_distribution["quantidade"] / class_distribution["quantidade"].sum()

display(class_distribution)

class_distribution.to_csv(
    OUTPUT_TABLES_DIR / "section_3_class_distribution_modeling.csv",
    index=False,
    encoding="utf-8"
)
```

```python
plt.figure(figsize=(6, 4))
plt.bar(class_distribution["classe"], class_distribution["quantidade"])
plt.title("Distribuição das classes de sentimento")
plt.xlabel("Classe")
plt.ylabel("Quantidade de documentos")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_3_class_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 3.3 Caracterização da tarefa supervisionada

A tarefa supervisionada desta seção consiste em prever a polaridade da avaliação a partir do texto lematizado.

A variável de entrada é:

- `text_lemma`: texto pré-processado da avaliação.

A variável alvo é:

- `sentiment_label`: classe de sentimento, com valores `positivo` ou `negativo`.

A distribuição das classes foi analisada antes do treinamento para verificar possível desbalanceamento, como avaliações reais de consumidores podem apresentar proporções diferentes entre comentários positivos e negativos, os modelos foram configurados com `class_weight="balanced"`, reduzindo o risco de favorecer apenas a classe majoritária.

```python
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42
TEST_SIZE = 0.20

X = df_model["text_lemma"]
y = df_model["sentiment_label"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
)

print("Quantidade de documentos no treino:", len(X_train))
print("Quantidade de documentos no teste:", len(X_test))

print("\nDistribuição no treino:")
display(y_train.value_counts(normalize=True).reset_index().rename(columns={"index": "classe", "sentiment_label": "proporcao"}))

print("\nDistribuição no teste:")
display(y_test.value_counts(normalize=True).reset_index().rename(columns={"index": "classe", "sentiment_label": "proporcao"}))
```

## 3.4 Modelos avaliados

Foram avaliados dois modelos supervisionados:

1. Regressão Logística;
2. SVM Linear.

Ambos foram treinados em pipelines do scikit-learn contendo:

- vetorização com TF-IDF;
- classificador supervisionado.

A representação TF-IDF foi reutilizada por ter apresentado bom comportamento na etapa de busca textual e por ser uma representação adequada para classificação de textos curtos e médios.

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

tfidf_params = {
    "max_features": 5000,
    "min_df": 3,
    "ngram_range": (1, 2)
}

models = {
    "Regressão Logística": Pipeline([
        ("tfidf", TfidfVectorizer(**tfidf_params)),
        ("clf", LogisticRegression(
            max_iter=1000,
            class_weight="balanced",
            random_state=RANDOM_STATE
        ))
    ]),
    "SVM Linear": Pipeline([
        ("tfidf", TfidfVectorizer(**tfidf_params)),
        ("clf", LinearSVC(
            class_weight="balanced",
            random_state=RANDOM_STATE,
            max_iter=5000
        ))
    ])
}

models
```

## 3.5 Treinamento e avaliação dos modelos

Os modelos serão comparados por meio das seguintes métricas:

- accuracy;
- precision macro;
- recall macro;
- F1 macro;
- precision weighted;
- recall weighted;
- F1 weighted.

A métrica F1 macro será usada como principal critério de comparação porque considera o desempenho médio entre as classes, sem favorecer apenas a classe mais frequente.

```python
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    classification_report,
    confusion_matrix
)

metrics_records = []
reports = {}
predictions = {}

for model_name, model in models.items():
    print("=" * 80)
    print("Treinando modelo:", model_name)
    print("=" * 80)
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    predictions[model_name] = y_pred
    
    precision_macro, recall_macro, f1_macro, _ = precision_recall_fscore_support(
        y_test,
        y_pred,
        average="macro",
        zero_division=0
    )
    
    precision_weighted, recall_weighted, f1_weighted, _ = precision_recall_fscore_support(
        y_test,
        y_pred,
        average="weighted",
        zero_division=0
    )
    
    accuracy = accuracy_score(y_test, y_pred)
    
    metrics_records.append({
        "modelo": model_name,
        "accuracy": accuracy,
        "precision_macro": precision_macro,
        "recall_macro": recall_macro,
        "f1_macro": f1_macro,
        "precision_weighted": precision_weighted,
        "recall_weighted": recall_weighted,
        "f1_weighted": f1_weighted
    })
    
    report = classification_report(
        y_test,
        y_pred,
        output_dict=True,
        zero_division=0
    )
    
    reports[model_name] = report
    
    print(classification_report(y_test, y_pred, zero_division=0))
```

```python
metrics_df = pd.DataFrame(metrics_records).sort_values("f1_macro", ascending=False)

display(metrics_df)

metrics_df.to_csv(
    OUTPUT_TABLES_DIR / "section_3_model_metrics.csv",
    index=False,
    encoding="utf-8"
)
```

```python
metrics_plot_df = metrics_df.set_index("modelo")[["precision_macro", "recall_macro", "f1_macro"]]

ax = metrics_plot_df.plot(kind="bar", figsize=(8, 5))

plt.title("Comparação dos modelos de classificação")
plt.xlabel("Modelo")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.legend(title="Métrica")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_3_model_metrics_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 3.6 Seleção do melhor modelo

O melhor modelo será selecionado com base no F1 macro, sendo essa métrica adequada porque combina precision e recall e trata as classes de forma equilibrada, o que é importante em tarefas de sentimento com possível desbalanceamento entre avaliações positivas e negativas.

```python
best_model_name = metrics_df.iloc[0]["modelo"]
best_model = models[best_model_name]
best_predictions = predictions[best_model_name]

print("Melhor modelo selecionado:", best_model_name)

best_metrics = metrics_df.iloc[0].to_dict()
best_metrics
```

```python
best_report_df = pd.DataFrame(reports[best_model_name]).transpose()

display(best_report_df)

best_report_df.to_csv(
    OUTPUT_TABLES_DIR / "section_3_best_model_classification_report.csv",
    index=True,
    encoding="utf-8"
)
```

## 3.7 Matriz de confusão do melhor modelo

A matriz de confusão permite observar os acertos e erros do melhor modelo em cada classe.

Essa visualização é importante porque mostra se o classificador está confundindo avaliações positivas com negativas ou se consegue separar adequadamente os dois grupos.

```python
from sklearn.metrics import ConfusionMatrixDisplay

class_labels = sorted(y_test.unique())

cm = confusion_matrix(
    y_test,
    best_predictions,
    labels=class_labels
)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=class_labels
)

fig, ax = plt.subplots(figsize=(6, 5))
disp.plot(ax=ax, values_format="d", colorbar=False)

plt.title(f"Matriz de confusão - {best_model_name}")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_3_best_model_confusion_matrix.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 4.1 Carregamento da base processada

A base utilizada nesta seção é `data/processed/repro_preprocessed.csv`, gerada na etapa de pré-processamento e serão utilizadas tanto informações textuais quanto informações estruturadas do corpus, como produto, marca, categoria, tópicos e polaridade.

```python
from pathlib import Path
import os
import re
import ast
import json
import unicodedata
from collections import Counter, defaultdict

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import spacy

BASE_DIR = Path.cwd()

DATA_PROCESSED_DIR = BASE_DIR / "data" / "processed"
OUTPUT_FIGURES_DIR = BASE_DIR / "outputs" / "figures"
OUTPUT_TABLES_DIR = BASE_DIR / "outputs" / "tables"
OUTPUT_GRAPHS_DIR = BASE_DIR / "outputs" / "graphs"

for directory in [OUTPUT_FIGURES_DIR, OUTPUT_TABLES_DIR, OUTPUT_GRAPHS_DIR]:
    directory.mkdir(parents=True, exist_ok=True)

processed_path = DATA_PROCESSED_DIR / "repro_preprocessed.csv"

df = pd.read_csv(processed_path)

print("Base carregada com sucesso.")
print("Dimensões:", df.shape)

display(df.head())
```

```python
required_columns = [
    "doc_id",
    "text_raw",
    "text_lemma",
    "product_name",
    "product_brand",
    "site_category_lv1",
    "site_category_lv2",
    "topics",
    "polarity"
]

missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"Colunas obrigatórias ausentes: {missing_columns}")

for col in required_columns:
    df[col] = df[col].fillna("").astype(str)

display(df[required_columns].head())
```

## 4.2 Preparação auxiliar dos rótulos e textos

Nesta etapa são criadas funções auxiliares para padronizar textos, categorias, marcas e polaridades.

A polaridade será simplificada em classes como `positivo`, `negativo`, `misto`, `neutro` e `outro`, permitindo relacionar padrões textuais e entidades do domínio com o sentimento das avaliações.

```python
def parse_list_like(value):
    if isinstance(value, (list, tuple, set)):
        return list(value)
    
    if pd.isna(value):
        return []
    
    value_str = str(value).strip()
    
    try:
        parsed = ast.literal_eval(value_str)
        if isinstance(parsed, (list, tuple, set)):
            return list(parsed)
        return [parsed]
    except Exception:
        return [value_str]

def simplify_polarity(value):
    values = {str(item).upper().strip() for item in parse_list_like(value)}
    
    has_positive = "POSITIVO" in values
    has_negative = "NEGATIVO" in values
    has_neutral = "NEUTRO" in values
    
    if has_positive and not has_negative and not has_neutral:
        return "positivo"
    
    if has_negative and not has_positive and not has_neutral:
        return "negativo"
    
    if has_neutral and not has_positive and not has_negative:
        return "neutro"
    
    if has_positive and has_negative:
        return "misto"
    
    return "outro"

def clean_label(value):
    value = "" if pd.isna(value) else str(value)
    value = re.sub(r"\s+", " ", value).strip()
    return value if value else "não informado"

def strip_accents(text):
    text = "" if pd.isna(text) else str(text)
    text = unicodedata.normalize("NFD", text)
    text = "".join(char for char in text if unicodedata.category(char) != "Mn")
    return text

df["sentiment_label"] = df["polarity"].apply(simplify_polarity)
df["category_lv1_clean"] = df["site_category_lv1"].apply(clean_label)
df["brand_clean"] = df["product_brand"].apply(clean_label)
df["product_clean"] = df["product_name"].apply(clean_label)
df["text_regex"] = df["text_raw"].str.lower().apply(strip_accents)

display(df[[
    "doc_id",
    "category_lv1_clean",
    "brand_clean",
    "sentiment_label",
    "text_raw"
]].head())
```

## 4.3 Extração de entidades nomeadas com spaCy

Nesta etapa, o modelo de português do spaCy é utilizado para extrair entidades nomeadas dos textos brutos das avaliações.

Embora avaliações de e-commerce sejam textos curtos e nem sempre contenham muitas entidades tradicionais, essa etapa ajuda a identificar possíveis organizações, locais, pessoas ou termos reconhecidos pelo modelo.

```python
nlp = spacy.load("pt_core_news_sm")

print("Modelo spaCy carregado.")
print("Pipeline:", nlp.pipe_names)
```

```python
ner_records = []

texts = df["text_raw"].fillna("").astype(str).tolist()
doc_ids = df["doc_id"].tolist()

for doc_id, doc in zip(doc_ids, nlp.pipe(texts, batch_size=128, n_process=1)):
    for ent in doc.ents:
        ent_text = clean_label(ent.text)
        
        if len(ent_text) > 2 and not ent_text.isdigit():
            ner_records.append({
                "doc_id": doc_id,
                "entity_text": ent_text,
                "entity_label": ent.label_
            })

ner_df = pd.DataFrame(ner_records)

print("Total de entidades extraídas:", len(ner_df))

display(ner_df.head(20))
```

```python
if len(ner_df) > 0:
    ner_type_distribution = (
        ner_df["entity_label"]
        .value_counts()
        .reset_index()
    )
    
    ner_type_distribution.columns = ["tipo_entidade", "frequencia"]
    
    top_entities = (
        ner_df.groupby(["entity_text", "entity_label"])
        .size()
        .reset_index(name="frequencia")
        .sort_values("frequencia", ascending=False)
        .head(30)
    )
else:
    ner_type_distribution = pd.DataFrame(columns=["tipo_entidade", "frequencia"])
    top_entities = pd.DataFrame(columns=["entity_text", "entity_label", "frequencia"])

display(ner_type_distribution)
display(top_entities)

ner_df.to_csv(
    OUTPUT_TABLES_DIR / "section_4_ner_entities.csv",
    index=False,
    encoding="utf-8"
)

ner_type_distribution.to_csv(
    OUTPUT_TABLES_DIR / "section_4_ner_type_distribution.csv",
    index=False,
    encoding="utf-8"
)

top_entities.to_csv(
    OUTPUT_TABLES_DIR / "section_4_top_ner_entities.csv",
    index=False,
    encoding="utf-8"
)
```

```python
if len(ner_type_distribution) > 0:
    plt.figure(figsize=(7, 4))
    plt.bar(ner_type_distribution["tipo_entidade"], ner_type_distribution["frequencia"])
    plt.title("Distribuição de entidades nomeadas extraídas com spaCy")
    plt.xlabel("Tipo de entidade")
    plt.ylabel("Frequência")
    plt.tight_layout()
    
    plt.savefig(
        OUTPUT_FIGURES_DIR / "section_4_ner_type_distribution.png",
        dpi=300,
        bbox_inches="tight"
    )
    
    plt.show()
else:
    print("Nenhuma entidade nomeada foi extraída.")
```

## 4.4 Extração de padrões de domínio com regex

Além das entidades nomeadas, foram definidos padrões textuais específicos do domínio de e-commerce.

Esses padrões representam temas recorrentes nas avaliações, como entrega, defeito, qualidade, preço, atendimento e divergência em relação ao anúncio, e essa abordagem é útil porque avaliações de produtos nem sempre contêm entidades nomeadas tradicionais, mas frequentemente contêm expressões de domínio relevantes para análise de negócio.

```python
domain_patterns = {
    "entrega_logistica": r"\b(entrega|entregar|chegou|chegar|prazo|atraso|atrasad\w*|demora\w*|recebi|receber|transportadora)\b",
    "defeito_funcionamento": r"\b(defeito|defeituos\w*|quebrad\w*|funciona\w*|funcionar|parou|troca|trocar|devolu\w*|garantia)\b",
    "qualidade_produto": r"\b(qualidade|excelente|otim\w*|bom|boa|ruim|pessim\w*|material|acabamento|resistente|fragil)\b",
    "preco_valor": r"\b(preco|valor|caro|cara|barato|barata|custo beneficio|beneficio|promocao|dinheiro)\b",
    "atendimento_suporte": r"\b(atendimento|suporte|sac|resposta|cliente|contato|resolver|solucao)\b",
    "anuncio_descricao": r"\b(anuncio|descricao|foto|imagem|diferente|divergente|propaganda|informacao)\b"
}

pattern_records = []

for _, row in df.iterrows():
    text = row["text_regex"]
    
    for pattern_name, pattern_regex in domain_patterns.items():
        matches = re.findall(pattern_regex, text)
        
        if matches:
            pattern_records.append({
                "doc_id": row["doc_id"],
                "pattern": pattern_name,
                "n_matches": len(matches),
                "sentiment_label": row["sentiment_label"],
                "category_lv1_clean": row["category_lv1_clean"],
                "brand_clean": row["brand_clean"]
            })

patterns_df = pd.DataFrame(pattern_records)

print("Total de ocorrências documento-padrão:", len(patterns_df))

display(patterns_df.head())
```

```python
pattern_distribution = (
    patterns_df
    .groupby("pattern")
    .agg(
        documentos=("doc_id", "nunique"),
        ocorrencias=("n_matches", "sum")
    )
    .reset_index()
    .sort_values("documentos", ascending=False)
)

display(pattern_distribution)

patterns_df.to_csv(
    OUTPUT_TABLES_DIR / "section_4_regex_pattern_records.csv",
    index=False,
    encoding="utf-8"
)

pattern_distribution.to_csv(
    OUTPUT_TABLES_DIR / "section_4_regex_pattern_distribution.csv",
    index=False,
    encoding="utf-8"
)
```

```python
plt.figure(figsize=(9, 5))
plt.barh(pattern_distribution["pattern"][::-1], pattern_distribution["documentos"][::-1])
plt.title("Padrões de domínio extraídos com regex")
plt.xlabel("Quantidade de documentos")
plt.ylabel("Padrão")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_4_regex_pattern_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 4.5 Construção do grafo de conhecimento

O grafo de conhecimento será construído com NetworkX a partir de relações extraídas do corpus.

Os nós do grafo representam:

- categorias de produto;
- marcas;
- padrões textuais de domínio;
- polaridades.

As arestas representam relações observadas nos documentos, como:

- categoria associada a marca;
- categoria associada a padrão textual;
- padrão textual associado a polaridade;
- marca associada a polaridade.

O peso das arestas indica a frequência com que a relação aparece no corpus.

```python
top_categories = set(
    df["category_lv1_clean"]
    .value_counts()
    .head(8)
    .index
)

top_brands = set(
    df.loc[df["brand_clean"] != "não informado", "brand_clean"]
    .value_counts()
    .head(12)
    .index
)

valid_sentiments = {"positivo", "negativo"}

patterns_by_doc = (
    patterns_df
    .groupby("doc_id")["pattern"]
    .apply(lambda values: sorted(set(values)))
    .to_dict()
)

G = nx.Graph()

def add_node_with_type(graph, node_id, label, node_type):
    if not graph.has_node(node_id):
        graph.add_node(
            node_id,
            label=label,
            node_type=node_type
        )

def add_weighted_edge(graph, source, target, relation):
    if graph.has_edge(source, target):
        graph[source][target]["weight"] += 1
    else:
        graph.add_edge(
            source,
            target,
            weight=1,
            relation=relation
        )

for _, row in df.iterrows():
    doc_id = row["doc_id"]
    category = row["category_lv1_clean"]
    brand = row["brand_clean"]
    sentiment = row["sentiment_label"]
    doc_patterns = patterns_by_doc.get(doc_id, [])
    
    if category not in top_categories:
        continue
    
    category_node = f"categoria::{category}"
    add_node_with_type(G, category_node, category, "categoria")
    
    if brand in top_brands:
        brand_node = f"marca::{brand}"
        add_node_with_type(G, brand_node, brand, "marca")
        add_weighted_edge(G, category_node, brand_node, "categoria_marca")
    
    if sentiment in valid_sentiments:
        sentiment_node = f"sentimento::{sentiment}"
        add_node_with_type(G, sentiment_node, sentiment, "sentimento")
        add_weighted_edge(G, category_node, sentiment_node, "categoria_sentimento")
        
        if brand in top_brands:
            add_weighted_edge(G, brand_node, sentiment_node, "marca_sentimento")
    
    for pattern in doc_patterns:
        pattern_node = f"padrao::{pattern}"
        add_node_with_type(G, pattern_node, pattern, "padrao")
        add_weighted_edge(G, category_node, pattern_node, "categoria_padrao")
        
        if sentiment in valid_sentiments:
            add_weighted_edge(G, pattern_node, sentiment_node, "padrao_sentimento")

print("Grafo criado.")
print("Quantidade de nós:", G.number_of_nodes())
print("Quantidade de arestas:", G.number_of_edges())

if G.number_of_nodes() < 20:
    raise ValueError("O grafo possui menos de 20 nós. Ajuste os limites de categorias ou marcas.")
```

```python
nodes_data = []

for node_id, attrs in G.nodes(data=True):
    nodes_data.append({
        "node_id": node_id,
        "label": attrs.get("label"),
        "node_type": attrs.get("node_type")
    })

edges_data = []

for source, target, attrs in G.edges(data=True):
    edges_data.append({
        "source": source,
        "target": target,
        "source_label": G.nodes[source].get("label"),
        "target_label": G.nodes[target].get("label"),
        "relation": attrs.get("relation"),
        "weight": attrs.get("weight")
    })

nodes_df = pd.DataFrame(nodes_data)
edges_df = pd.DataFrame(edges_data).sort_values("weight", ascending=False)

display(nodes_df.head())
display(edges_df.head(20))

nodes_df.to_csv(
    OUTPUT_TABLES_DIR / "section_4_graph_nodes.csv",
    index=False,
    encoding="utf-8"
)

edges_df.to_csv(
    OUTPUT_TABLES_DIR / "section_4_graph_edges.csv",
    index=False,
    encoding="utf-8"
)
```

## 4.6 Centralidade do grafo

A centralidade permite identificar os nós mais importantes na estrutura relacional.

Serão calculadas três medidas:

- grau: quantidade de conexões de um nó;
- grau ponderado: soma dos pesos das conexões;
- centralidade de intermediação: mede o quanto um nó atua como ponte entre diferentes partes do grafo.

Essas métricas ajudam a interpretar quais categorias, marcas, padrões ou polaridades são mais centrais no domínio analisado.

```python
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G, weight="weight", normalized=True)

centrality_records = []

for node_id, attrs in G.nodes(data=True):
    weighted_degree = sum(
        edge_attrs.get("weight", 1)
        for _, _, edge_attrs in G.edges(node_id, data=True)
    )
    
    centrality_records.append({
        "node_id": node_id,
        "label": attrs.get("label"),
        "node_type": attrs.get("node_type"),
        "degree": G.degree(node_id),
        "weighted_degree": weighted_degree,
        "degree_centrality": degree_centrality.get(node_id, 0),
        "betweenness_centrality": betweenness_centrality.get(node_id, 0)
    })

centrality_df = (
    pd.DataFrame(centrality_records)
    .sort_values(["weighted_degree", "degree_centrality"], ascending=False)
)

display(centrality_df.head(20))

centrality_df.to_csv(
    OUTPUT_TABLES_DIR / "section_4_graph_centrality.csv",
    index=False,
    encoding="utf-8"
)
```

```python
top_central_nodes = centrality_df.head(15).copy()

plt.figure(figsize=(10, 6))
plt.barh(top_central_nodes["label"][::-1], top_central_nodes["weighted_degree"][::-1])
plt.title("Nós mais centrais do grafo por grau ponderado")
plt.xlabel("Grau ponderado")
plt.ylabel("Nó")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_4_top_central_nodes.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 4.7 Visualização do grafo

Para facilitar a interpretação visual, será exibido um subgrafo com os nós mais centrais, essa visualização mostra como categorias, marcas, padrões e polaridades se conectam no domínio analisado.

```python
top_nodes_for_viz = set(centrality_df.head(40)["node_id"])
G_viz = G.subgraph(top_nodes_for_viz).copy()

node_type_colors = {
    "categoria": "#1f77b4",
    "marca": "#ff7f0e",
    "padrao": "#2ca02c",
    "sentimento": "#d62728"
}

node_colors = [
    node_type_colors.get(G_viz.nodes[node].get("node_type"), "#7f7f7f")
    for node in G_viz.nodes()
]

node_sizes = [
    300 + (centrality_df.set_index("node_id").loc[node, "degree_centrality"] * 5000)
    for node in G_viz.nodes()
]

edge_widths = [
    0.5 + np.log1p(G_viz[u][v].get("weight", 1))
    for u, v in G_viz.edges()
]

pos = nx.spring_layout(G_viz, seed=42, k=0.8)

plt.figure(figsize=(14, 10))

nx.draw_networkx_edges(
    G_viz,
    pos,
    width=edge_widths,
    alpha=0.35
)

nx.draw_networkx_nodes(
    G_viz,
    pos,
    node_color=node_colors,
    node_size=node_sizes,
    alpha=0.9
)

labels = {
    node: G_viz.nodes[node].get("label")
    for node in G_viz.nodes()
}

nx.draw_networkx_labels(
    G_viz,
    pos,
    labels=labels,
    font_size=8
)

legend_handles = [
    plt.Line2D(
        [0],
        [0],
        marker="o",
        color="w",
        label=node_type,
        markerfacecolor=color,
        markersize=10
    )
    for node_type, color in node_type_colors.items()
]

plt.legend(handles=legend_handles, title="Tipo de nó", loc="best")
plt.title("Grafo de conhecimento: categorias, marcas, padrões e sentimentos")
plt.axis("off")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_4_knowledge_graph.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 4.8 Pergunta analítica respondida pelo grafo

Pergunta de negócio:

**Quais padrões textuais aparecem mais associados a avaliações negativas e que tipos de problema eles revelam sobre a experiência do consumidor?**

Para responder a essa pergunta, serão analisadas as conexões entre padrões extraídos por regex e a polaridade negativa.

```python
negative_patterns = (
    patterns_df[patterns_df["sentiment_label"] == "negativo"]
    .groupby("pattern")
    .agg(
        documentos_negativos=("doc_id", "nunique"),
        ocorrencias_negativas=("n_matches", "sum")
    )
    .reset_index()
)

all_patterns = (
    patterns_df
    .groupby("pattern")
    .agg(
        documentos_total=("doc_id", "nunique"),
        ocorrencias_total=("n_matches", "sum")
    )
    .reset_index()
)

negative_pattern_summary = (
    all_patterns
    .merge(negative_patterns, on="pattern", how="left")
    .fillna(0)
)

negative_pattern_summary["proporcao_documentos_negativos"] = (
    negative_pattern_summary["documentos_negativos"]
    / negative_pattern_summary["documentos_total"]
)

negative_pattern_summary = negative_pattern_summary.sort_values(
    ["documentos_negativos", "proporcao_documentos_negativos"],
    ascending=False
)

display(negative_pattern_summary)

negative_pattern_summary.to_csv(
    OUTPUT_TABLES_DIR / "section_4_negative_pattern_summary.csv",
    index=False,
    encoding="utf-8"
)
```

```python
negative_category_pattern = (
    patterns_df[patterns_df["sentiment_label"] == "negativo"]
    .groupby(["category_lv1_clean", "pattern"])
    .agg(
        documentos_negativos=("doc_id", "nunique"),
        ocorrencias_negativas=("n_matches", "sum")
    )
    .reset_index()
    .sort_values("documentos_negativos", ascending=False)
    .head(20)
)

display(negative_category_pattern)

negative_category_pattern.to_csv(
    OUTPUT_TABLES_DIR / "section_4_negative_category_pattern.csv",
    index=False,
    encoding="utf-8"
)
```

```python
plt.figure(figsize=(9, 5))
plt.barh(
    negative_pattern_summary["pattern"][::-1],
    negative_pattern_summary["documentos_negativos"][::-1]
)
plt.title("Padrões mais associados a avaliações negativas")
plt.xlabel("Quantidade de documentos negativos")
plt.ylabel("Padrão")
plt.tight_layout()

plt.savefig(
    OUTPUT_FIGURES_DIR / "section_4_negative_patterns.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 5.3 Tabela final de indicadores do projeto

```python
final_indicators = []

metadata_path = DATA_INTERIM_DIR / "metadata_repro.json"
if metadata_path.exists():
    with open(metadata_path, "r", encoding="utf-8") as f:
        metadata = json.load(f)
    
    final_indicators.append({
        "dimensao": "Corpus",
        "indicador": "Quantidade de documentos",
        "valor": metadata.get("n_documentos")
    })

vocab_impact = read_csv_if_exists(OUTPUT_TABLES_DIR / "section_1_vocab_impact.csv")
if not vocab_impact.empty:
    lemma_vocab = vocab_impact.loc[
        vocab_impact["representacao"] == "lemmatizacao",
        "tamanho_vocabulario"
    ]
    
    if len(lemma_vocab) > 0:
        final_indicators.append({
            "dimensao": "Pré-processamento",
            "indicador": "Vocabulário lematizado",
            "valor": int(lemma_vocab.iloc[0])
        })

vectorization_summary = read_csv_if_exists(OUTPUT_TABLES_DIR / "section_2_vectorization_summary.csv")
if not vectorization_summary.empty:
    tfidf_terms = vectorization_summary.loc[
        vectorization_summary["representacao"] == "TF-IDF",
        "termos"
    ]
    
    if len(tfidf_terms) > 0:
        final_indicators.append({
            "dimensao": "Representação vetorial",
            "indicador": "Termos na matriz TF-IDF",
            "valor": int(tfidf_terms.iloc[0])
        })

model_metrics = read_csv_if_exists(OUTPUT_TABLES_DIR / "section_3_model_metrics.csv")
if not model_metrics.empty:
    best_model_row = model_metrics.sort_values("f1_macro", ascending=False).iloc[0]
    
    final_indicators.append({
        "dimensao": "Classificação",
        "indicador": "Melhor modelo",
        "valor": best_model_row["modelo"]
    })
    
    final_indicators.append({
        "dimensao": "Classificação",
        "indicador": "F1 macro do melhor modelo",
        "valor": round(best_model_row["f1_macro"], 4)
    })

graph_nodes_path = OUTPUT_TABLES_DIR / "section_4_graph_nodes.csv"
graph_edges_path = OUTPUT_TABLES_DIR / "section_4_graph_edges.csv"

graph_nodes = read_csv_if_exists(graph_nodes_path)
graph_edges = read_csv_if_exists(graph_edges_path)

if not graph_nodes.empty:
    final_indicators.append({
        "dimensao": "Grafo",
        "indicador": "Quantidade de nós",
        "valor": len(graph_nodes)
    })

if not graph_edges.empty:
    final_indicators.append({
        "dimensao": "Grafo",
        "indicador": "Quantidade de arestas",
        "valor": len(graph_edges)
    })

final_indicators_df = pd.DataFrame(final_indicators)

display(final_indicators_df)

final_indicators_df.to_csv(
    OUTPUT_TABLES_DIR / "section_5_final_indicators.csv",
    index=False,
    encoding="utf-8"
)
```

## 5.6 Síntese interpretativa para stakeholder não técnico

A síntese abaixo apresenta os principais achados do projeto em linguagem acessível, voltada a um público não técnico, o objetivo é comunicar o que foi descoberto a partir das avaliações, sem focar apenas nos detalhes das técnicas utilizadas.

As avaliações de consumidores foram analisadas para identificar padrões de satisfação e insatisfação em compras de e-commerce. O pipeline mostrou que os textos das avaliações contêm sinais suficientes para distinguir automaticamente comentários positivos e negativos, além de revelar quais temas aparecem com maior frequência nas experiências dos consumidores.

A análise indicou que os consumidores falam principalmente sobre qualidade do produto, entrega, preço, defeitos, atendimento e diferenças entre o anúncio e o produto recebido. Entre esses temas, qualidade do produto e entrega logística apareceram como fatores muito relevantes no corpus.

O modelo de classificação de sentimento conseguiu separar avaliações positivas e negativas com bom desempenho, isso indica que, a partir do texto escrito pelo consumidor, é possível automatizar parte da triagem de opiniões, identificando rapidamente avaliações de satisfação e reclamações.

O grafo de conhecimento mostrou como categorias, marcas, padrões textuais e sentimentos se conectam, e os resultados indicaram que problemas de entrega, qualidade do produto e defeito ou funcionamento aparecem fortemente associados a avaliações negativas, oque sugere que melhorias na logística, na descrição dos produtos e no controle de qualidade podem ter impacto direto na experiência do consumidor.

Em termos práticos, o projeto demonstra que avaliações textuais podem ser transformadas em informações úteis para apoiar decisões de negócio, como priorizar categorias críticas, monitorar reclamações recorrentes e identificar pontos de melhoria na jornada de compra.

## 5.7 Principais achados técnicos

Os principais achados técnicos do pipeline foram:

1. A lematização foi mais adequada que o stemming como representação principal, pois preservou melhor a interpretabilidade dos termos.
2. A representação TF-IDF com unigramas e bigramas funcionou bem para busca textual e classificação.
3. O motor de busca por similaridade retornou documentos coerentes para consultas relacionadas a defeito, entrega e custo-benefício.
4. O SVM Linear apresentou melhor desempenho na classificação binária de sentimento.
5. A extração de padrões de domínio com regex foi mais útil para análise de negócio do que apenas o NER genérico.
6. O grafo de conhecimento permitiu relacionar categorias, marcas, padrões textuais e sentimentos, revelando temas centrais no corpus.

## 5.8 Limitações do pipeline

Apesar dos bons resultados, o pipeline possui algumas limitações.

A primeira limitação está relacionada ao tamanho e ao tipo dos textos, avaliações de e-commerce tendem a ser curtas, informais e, em alguns casos, ambíguas, oque pode dificultar tarefas como lematização, classificação de sentimento e extração de entidades.

A segunda limitação está na classificação binária de sentimento, para tornar a tarefa mais clara, avaliações neutras e mistas foram removidas da etapa supervisionada. Em uma aplicação real, seria importante tratar também esses casos, pois muitos consumidores expressam elogios e reclamações no mesmo comentário.

A terceira limitação está no uso de padrões regex que embora os padrões definidos sejam interpretáveis, eles dependem de termos escolhidos manualmente, assim algumas variações linguísticas podem não ser capturadas.

A quarta limitação está no NER genérico do spaCy, como o modelo não foi treinado especificamente para e-commerce, algumas marcas, produtos ou termos de domínio podem não ser reconhecidos corretamente.

Por fim, o grafo de conhecimento representa associações observadas no corpus, mas não relações causais. Portanto, os resultados devem ser interpretados como indícios de coocorrência e recorrência, não como prova de causalidade.

## 5.9 Melhorias futuras

Como melhorias futuras, seria possível:

1. Expandir a classificação de sentimento para múltiplas classes, incluindo avaliações neutras e mistas.
2. Testar modelos baseados em embeddings contextualizados em português, como BERTimbau, caso houvesse maior disponibilidade computacional.
3. Criar um modelo NER customizado para reconhecer marcas, produtos e termos específicos de e-commerce.
4. Refinar os padrões regex com base em análise manual de erros e novas expressões frequentes.
5. Construir um painel interativo para explorar categorias, marcas, sentimentos e padrões de reclamação.
6. Avaliar o pipeline em outro corpus de avaliações para testar sua generalização.

## 5.10 Conclusão geral do pipeline

O projeto construiu um pipeline completo de Processamento de Linguagem Natural aplicado a avaliações de produtos em português.

O pipeline partiu de um corpus real, realizou pré-processamento textual, construiu representações vetoriais, implementou busca textual por similaridade, treinou modelos de classificação de sentimento, extraiu entidades e padrões de domínio, construiu um grafo de conhecimento e consolidou visualizações interpretáveis.

Os resultados mostram que textos de avaliações de e-commerce podem ser transformados em informação estruturada e útil, e o pipeline permitiu identificar padrões recorrentes de satisfação e insatisfação, classificar sentimentos e visualizar relações entre categorias, marcas, temas e polaridades.
